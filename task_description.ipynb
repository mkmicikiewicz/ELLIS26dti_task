## Data Description

The dataset (based mostly on [1] and, as such, on [2] as well) contains protein–ligand binding affinity measurements at the level of unique protein–ligand pairs.

- **uniprot_id**: UniProt accession identifying the protein target. 
- **protein**: *Full* protein sequence obtained from UniProt. For example: Rows with the 'uniprot_id' of 'Q70CQ3[57-517]' will have the same 'protein' as 'Q70CQ3'. However, the mutations and their locations (as indicated in the 'uniprot_id') are carried over
- **clean_smiles**: Clean, canonical smiles after standardization. For each protein in the dataset, there are at least 10 unique ligands
- **pIC50**: binding affinity value
